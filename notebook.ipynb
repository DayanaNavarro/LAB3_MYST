# <font color= #6B6B6B> <font size = 6> 1. Introduction </font>

<hr style="border:0.02in solid gray"> </hr>

El Behavioral Finance lo investigaron en los años 70 los psicólogos Daniel Kahneman y Amos Tversky, autores del artículo ‘Prospect Theory: An Analysis of Decision under Risk’. Ellos plantearon con evidencia empírica que los seres humanos son irracionales de forma consistente.

Más adelante, los economistas Robert Shiller y Richard Thaler ganaron el premio Nobel de Economía gracias a sus hallazgos en esta área de investigación.

A lo largo de nuestra vida debemos tomar decisiones económicas, de mayor o menor trascendencia, pero todas influenciadas por varios elementos. Como las prioridades, el tiempo de espera y los beneficios a obtener; sin embargo no todo se basa en la racionalidad. 

La teoría de Behavioral Finance ha reforzado que las emociones y el aspecto psicológico de los individuos tiene un peso importante en las decisiones financieras que toman, creando con esto sesgos psiológicos que afectan en los mercados financieros.

A pesar de que las finanzas conductuales tienen su origen en la psicología, sus implicaciones están tan ligadas al mundo económico que estudian ambos círculos académicos. 

Anteriormente se creía que los agentes económicos eran racionales, pero el riesgo e incertidumbre influyen en los inversionistas y eso se refleja en los mercados y como esto parece incidir en los precios de los mercados financieros.Y por qué los mercados son tan proclives a periodos de volatilidad o a la formación de burbujas en el precio de los activos.

Esta teoría se ha vuelto tan relevante para el análisis bursátil porque la euforia colectiva puede provocar grandes caídas o alzas en los precios. Aparentemente, no solo las noticias financieras o políticas tienen un impacto relevante en el mercado bursátil. 

Las finanzas conductuales ayudan a entender que las decisiones financieras en cuanto a inversiones, pagos, riesgos y deudas personales tienen mucha influencia emocional, sesgos y relación con el contexto de los seres humanos.

Esa comprensión ayuda a tomar mejores decisiones y a ser más racionales en los asuntos financieros; a analizar la lógica y eficiencia de los mercados, y sus distintos escenarios.

Definiendo lo anterior se explicaran algunos otros términos mencionados como:

Los pips que son el cambio mínimo posible de precio de un activo financiero. Para los pares de divisas, este es, generalmente, el cuarto dígito después del punto decimal.

El radio de sharpe original y actualizado,se utiliza para mostrar hasta qué punto el rendimiento de una inversión compensa al inversor por asumir riesgo en su inversión, es un buen método para medir la desviación standard de la rentabilidad de cualquier fondo individualmente y compararla con otros.

Los DrawDown y DrawUp, un drawdown se refiere a cuánto ha caído una cuenta desde su punto máximo hasta su punto mínimo en términos de capital o monto de inversión y el drawup se refiere a cuánto ha subido una cuenta desde su punto mínimo hasta su punto máximo en términos de capital o monto de inversión.
